```python
                            #Task 1

# Dictionary containing student names as keys and their marks as values
students = {"Zayn": 90, "usman": 80, "waheed": 70}

# Function to print each student's result and grade
def print_result(data):
    for name, marks in data.items():
        # Assign grade based on marks
        if marks >= 80:
            grade = 'A'
        elif marks >= 60:
            grade = 'B'
        elif marks >= 50:
            grade = 'C'
        else:
            grade = 'fail'

        # Print the student's name, marks, and grade
        print(f"{name}: {marks} -> Grade: {grade}")

# Call the function to print results for all students
print_result(students)

# Calculate the average marks of the class
marks = students.values()
average = sum(marks) / len(marks)
print(f"And average of marks in class is {average}")
```

```python

                                #Task 4
import random

# Dictionary containing movie genres as keys and lists of movies as values
movies = {
    "action": ["Avengers", "Batman", "Spiderman"],
    "comedy": ["Mr. Bean", "Home Alone", "The Mask"],
    "sci-fi": ["Interstellar", "Inception", "The Matrix"],
    "horror": ["Conjuring", "Annabelle", "It"]
}

# Prompt the user to enter a genre and convert input to lowercase
genre = input("Enter genre (action/comedy/sci-fi/horror): ").lower()

# Check if the entered genre exists in the movies dictionary
if genre in movies:
    # Randomly select and print a movie from the chosen genre
    print("Watch:", random.choice(movies[genre]))
else:
    # Print an error message if the genre is not found
    print("Sorry! try again: action, sci-fi, comedy, or horror")
```

```python
                                        #Task 6

numbers = []

# Get numbers from user as a single line of input
user_input = input("Enter numbers separated by space: ")

# Make a list of integers from the input
numbers = [int(n) for n in user_input.split()]

def analyze(num):
    # Find the largest number
    maximum = max(num)
    # Find the smallest number
    minimum = min(num)
    # Find the average of the numbers
    average = sum(num) / len(num)
    
    print("numbers:", num)
    print("max:", maximum)
    print("min:", minimum)
    print("average:", average)
    
    # Make a list of numbers greater than the average
    above_avg = [n for n in num if n > average]
    print("above average:", above_avg)

# Call the function to show results
analyze(numbers)
```

```python
students = {}

# Get how many students from user
n = int(input("Enter number of students: "))

# Get each student's name and marks from user
for i in range(n):
    name = input(f"Enter name of student {i+1}: ")
    marks = int(input(f"Enter marks of {name}: "))
    students[name] = marks

def report(data):
    for name, marks in data.items():
        # Decide grade and result
        if marks >= 80:
            grade = 'A'
            result = "pass"
        elif marks >= 60:
            grade = 'B'
            result = "pass"
        elif marks >= 50:
            grade = 'C'
            result = "pass"
        else:
            grade = 'F'
            result = "fail"

        # Show student info
        print(f"{name}\t{marks}\t{grade}\t{result}")

    # Find and show class average and topper
    avg = sum(data.values()) / len(data)
    topper = max(data, key=data.get)
    print("\nClass average:", round(avg, 2))
    print("Topper:", topper, "with", data[topper], "marks")

# Show the report for all students
report(students)
```
